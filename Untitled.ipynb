```python
#import les packages
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
files=[file for file in os.listdir(r'C:\Users\HP\Downloads\Sales_Data')]
for file in files:
    print(file)
```

```python
path=r'C:\Users\HP\Downloads\Sales_Data'
#cree une  base de donnee vide
all_data=pd.DataFrame()
for file in files:
    current_data=pd.read_csv(path+'/'+file)
    all_data=pd.concat([all_data,current_data])
print(all_data)
```

### Repondre a la premier question quel est le mois durant lequen on realise plus de vente donc l'idee est de voir le chaifre d'affaire generé dans chaque mois????

##### fuction pour trouver le mois 
def month(x):
    return x.split('/')[0]


```python
all_data['Month']=all_data['Order Date'].apply(month)
all_data
```

```python
all_data['Month'].unique()
```

```python
all_data=all_data[all_data['Month']!='Order Date']
all_data['Month'].unique()
```

```python
all_data.dtypes
```

```python
all_data["Month"]=all_data["Month"].astype(int)
all_data["Price Each"]=all_data["Price Each"].astype(float)
all_data["Quantity Ordered"]=all_data["Quantity Ordered"].astype(int)
all_data.dtypes
```

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
```

```python
all_data.groupby('Month')['Sales'].sum()
```

```python
months=range(1,13)
plt.bar(months,all_data.groupby('Month')['Sales'].sum())
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Month number')
plt.show()
```

### En quelle moment doit on faire une compagne publicitaire pour avoir plus de ventes?

```python
all_data['Hour']=pd.to_datetime(all_data['Order Date']).dt.hour
```

```python
all_data
```

```python
keys=[]
hours=[]
for key,hour in all_data.groupby('Hour'):
    keys.append(key)
    hours.append(len(hour))
```

```python
all_data
```

```python
plt.grid()
plt.plot(keys,hours)
plt.xlabel('heure de la journee')
plt.ylabel('nombre des commandes')
```
